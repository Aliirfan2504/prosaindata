```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
#pemangilan list
tweets_list=[]
```

```python
for i,tweet in enumerate(sntwitter.TwitterSearchScraper("Anis Baswedan since:2022-12-01 until:2023-02-19").get_items()):
    if i>1000:
        break
    tweets_list.append([tweet.content,tweet.username,])
```

```python
tweet = pd.DataFrame(tweets_list,columns=["text","username"])
```
